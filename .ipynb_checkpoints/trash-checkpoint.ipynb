```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import tensorflow.keras as tf
import os
import cv2
```

```python
from tensorflow.keras.preprocessing import image
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
```

```python
DIR = os.getcwd() + "\\trash_img"
CATEGORIES = [c for c in os.listdir(DIR)]

for CAT in CATEGORIES:
    PATH = os.path.join(DIR, CAT)
    for img in os.listdir(PATH):
        img_arr = cv2.imread(os.path.join(PATH, img))
        img_arr = cv2.resize(img_arr, (224, 224))
        img_arr = image.img_to_array(img_arr)
        img_arr = img_arr / 255
        plt.imshow(img_arr, cmap="gray")
        plt.show()
        break
    break
```

```python
from random import shuffle

# skrócić img, ponieważ datasety są za duże (2500+ img)
TO_CHANGE = ["organic", "plastic", "paper"]

X = []
for CH in TO_CHANGE:
    PATH = os.path.join(DIR, CH)
    for img in os.listdir(PATH):
        X.append(img)
    shuffle(X)
    X = X[:1700]
    for img in os.listdir(PATH):
        if img not in X:
            os.remove(os.path.join(PATH, img))
```

```python
X, y = [], []

for CAT in CATEGORIES:
    PATH = os.path.join(DIR, CAT)
    label = CATEGORIES.index(CAT)
    for img in os.listdir(PATH):
        img_arr = cv2.imread(os.path.join(PATH, img))
        img_arr = cv2.resize(img_arr, (224, 224))
        img_arr = image.img_to_array(img_arr)
        img_arr = img_arr / 255
        X.append(img_arr)
        y.append(label)
```

```python
X = np.array(X)
y = to_categorical(y)
```

```python
train_x, test_x, train_y, test_y = train_test_split(X, y, random_state=43, shuffle=True, test_size=0.2)

aug = image.ImageDataGenerator(
            rotation_range = 20,
            zoom_range=0.15,
            width_shift_range=0.2,
            height_shift_range=0.2,
            shear_range=0.15,
            horizontal_flip=True,
            fill_mode="nearest")
```
